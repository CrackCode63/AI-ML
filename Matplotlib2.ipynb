```python
import matplotlib.pyplot as plt
x = ["tum","padayi","khana","apna khayal"]
y = [80,10,8,2]
# c = ["red","green","yellow","blue"]
explode = [0.0,0.0,0.0,0.1]
plt.pie(y,labels=x,shadow= True, explode = explode)
plt.show()
```

```python
import matplotlib.pyplot as plt
x = ["tum","padayi","khana","apna khayal"]
y = [80,10,8,2]
c= ["r","g","y","black"]
explode = [0.2,0.0,0.0,0.0]
plt.pie(y,labels=x,shadow= True,radius= 1,explode = explode,colors= c)
plt.pie([])
plt.show()
```

```python
import matplotlib.pyplot as plt
x = [1,2,3,4,5]
area1 = [2,3,4,5,6]
area2 = [3,2,5,6,1]
area3 = [2,4,1,4,2]
l = ["area1","area2","area3"]
plt.stackplot(x,area1,area2,area3, labels = l)
plt.title("mera h")
plt.xlabel("x-axis")
plt.ylabel("y-axis")
plt.legend(loc= 2)
# legend fun is used to denotes the color of the region.
# It takes location as arguments.
plt.show()
```

```python
import matplotlib.pyplot as plt
x = [1,2,3,4,5,6,7]
y = [2,1,3,5,3,6,4]
c = [10, 20, 30, 40, 50, 60, 70]
s = [100,150,200,250,300,350,400]
plt.xlabel("language")
plt.ylabel("city")
plt.title("language and its city")
scatter = plt.scatter(x,y, c = c, cmap = "viridis" , marker = "*")
colorbar = plt.colorbar(scatter)
colorbar.set_label("colorbar",fontsize = 15)

plt.show()
```
