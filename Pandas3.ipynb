```python
# concatenating the dataframes
import pandas as pd
d1 = pd.DataFrame({"A":[1,2,3,4],"B":[11,12,13,14]})
d2 = pd.DataFrame({"A":[1,2,3,5],"B":[21,22,23,24]})
pd.concat([d1,d2])
# Here concat() is fun is used to cancatenate the dataframes.
# It takes dataframes name as argument
# It has an attribute named axix. The value of this attribute is 0 and 1.
```

```python
import pandas as pd
d1 = pd.DataFrame({"A":[1,2,3,4],"B":[11,12,13,14]})
d2 = pd.DataFrame({"A":[1,2],"B":[21,22]})
pd.concat([d1,d2],axis = 1, join = "outer")
# It has another attribute named join.
# The value of this attribute is outer, inner.
```

```python
import pandas as pd
d1 = pd.DataFrame({"A":[1,2,3,4],"B":[11,12,13,14]})
d2 = pd.DataFrame({"A":[1,2],"B":[21,22]})
pd.concat([d1,d2],axis = 1, join = "inner")
```

```python
# Key parameters in pandas
import pandas as pd
d1 = pd.DataFrame({"A":[1,2,3,4],"B":[11,12,13,14]})
d2 = pd.DataFrame({"A":[1,2,3,4],"C":[21,22,23,24]})
pd.concat([d1,d2],axis = 1,keys = ["d1","d2"])
```

```python
# Key parameters in pandas
import pandas as pd
d1 = pd.DataFrame({"A":[1,2,3,4],"B":[11,12,13,14]})
d2 = pd.DataFrame({"A":[1,2,3,4],"B":[21,22,23,24]})
pd.concat([d1,d2],keys = ["d1","d2"])
```

```python
# Key parameters in pandas
import pandas as pd
d1 = pd.DataFrame({"A":[1,2,3,4],"B":[11,12,13,14]})
d2 = pd.DataFrame({"C":[1,2,3,4],"D":[21,22,23,24]})
pd.concat([d1,d2],keys = ["d1","d2"])
```

```python
# Group By in pandas
import pandas as pd
d1 = pd.DataFrame({"Name":["A","B","C","D","A","B","E"],"s1":[12,13,14,12,13,14,15],"s2":[12,13,14,12,13,14,15]})
d2 = d1.groupby("Name")
d2
for x,y in d2:
    print(x)
    print(y)
    print()
```

```python
  # Melt func
import pandas as pd
d1 = pd.DataFrame({"days":[1,2,3,4,5,6],"eng":[10,20,14,13,15,16],"math":[14,15,12,15,16,13]})
pd.melt(d1)
```

```python
# pivot function is used to reshape 
import pandas as pd
d1 = pd.DataFrame({"days":[1,2,3,4,5,6],"eng":[10,20,14,13,15,16],"math":[14,15,12,15,16,13],"st_name":["ABHI","KIRAN","ABHI","KIRAN","ABHI","KIRAN"]})
d1.pivot(index = "days", columns = "st_name")
```
