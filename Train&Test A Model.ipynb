```python
# seperate data in dependant and independent values
# x represent independent values
x = dataset.iloc[:,1:-1]
x
```

```python
# Seperate dependent values
# y represent dependent values
y = dataset.iloc[:,5]
y
```

```python
from sklearn.preprocessing import LabelEncoder
labelEncoder = LabelEncoder()
# y = y.apply(labelEncoder.fit_transform)
# This can be written as
y = labelEncoder.fit_transform(y)
y
```

```python
from sklearn.preprocessing import LabelEncoder
labelEncoder = LabelEncoder()
x = x.apply(labelEncoder.fit_transform)
x
```

```python
# split data in training and testing
from sklearn.model_selection import train_test_split
x_train, x_test, y_train, y_test = train_test_split(x, y, test_size = 0.9, random_state = 0)
```

```python
# Apply Decision Tree Classifier
from sklearn.tree import DecisionTreeClassifier
DTC = DecisionTreeClassifier()
DTC.fit(x_train, y_train)
```

```python
# Checking the accuracy score of the model
from sklearn.metrics import accuracy_score

y_pred = DTC.predict(x_test)
accuracy = accuracy_score(y_test, y_pred)
print(accuracy)
```

```python
# Check prediction for new entry
z = np.array([1,1,0,1])
abhi = DTC.predict([z])
print(abhi)
```
